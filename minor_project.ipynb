```python
import pandas as pd
import matplotlib.pyplot as plt

history = pd.read_csv("training_history.csv")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(history["epoch"], history["train_loss"], label="Train loss")
ax1.plot(history["epoch"], history["val_loss"],   label="Val loss")
ax1.set_title("Loss over epochs")
ax1.set_xlabel("Epoch")
ax1.legend()

ax2.plot(history["epoch"], history["train_f1"], label="Train F1")
ax2.plot(history["epoch"], history["val_f1"],   label="Val F1")
ax2.set_title("F1 score over epochs")
ax2.set_xlabel("Epoch")
ax2.legend()

plt.tight_layout()
plt.savefig("training_curves.png", dpi=150)
plt.show()

from google.colab import files
files.download("training_curves.png")
```
